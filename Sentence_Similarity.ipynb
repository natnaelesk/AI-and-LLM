```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
sentences1 = ["i love to code", "let us walk outside" , "i didn't like last time we go to that cafe"]
sentences2 = ["have you seen the new movie", "i want to walk outside" , "last time we went to this cafe"]

emb1 = model.encode(sentences1)
emb2 = model.encode(sentences2)
```

```python
print(emb1)
print(emb2)
```

```python
from sentence_transformers import util
```

```python
cos_val = util.cos_sim(emb1,emb2)
print(cos_val)
```

```python
for i in range(len(cos_val)):
    print (" {} \t\t {} \n Score: {:.1f}%\n".format(sentences1[i],sentences2[i],(cos_val[i][i] * 100)))
```
