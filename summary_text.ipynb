```python
from transformers import pipeline
```

```python
summerizer =  pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
text = """
Artificial intelligence (AI) has been a topic of intense research and development for several decades. However, in recent years, AI has reached a point where it is no longer confined to the realms of science fiction or specialized industries. The applications of AI are now deeply embedded in our daily lives, from the way we communicate to the products we use and even the decisions we make. In healthcare, AI systems are being used to analyze medical data, diagnose conditions, and recommend treatments. In finance, AI-driven algorithms are used for stock trading and fraud detection. In manufacturing, robots and AI systems have revolutionized production lines, making them more efficient and reducing human labor.

As AI becomes more pervasive, concerns have arisen about its potential impact on the job market. Many jobs that were once done by humans are now being replaced by machines, leading to fears of mass unemployment. While AI certainly offers numerous benefits, it also presents ethical challenges. The idea of AI making decisions on behalf of humans, especially in critical areas like healthcare and law enforcement, raises concerns about accountability, transparency, and bias. Moreover, the increasing reliance on AI systems could create vulnerabilities, particularly in the case of system failures or malicious attacks.

Despite these challenges, there is an ongoing effort to develop responsible AI systems that are designed to complement human work rather than replace it. Governments, companies, and researchers are working together to establish ethical guidelines and policies for the development and deployment of AI technologies. The future of AI remains uncertain, but its potential to reshape society is undeniable.
"""
```

```python
response = summerizer(text)
```
