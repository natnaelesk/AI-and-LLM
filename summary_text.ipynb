 Drop in a big chunk of English text, and this mini code will crunch it down to a neat little summary. Perfect when you’re too lazy to read the whole thing 😅

i used the **BART large CNN model** from 🤗 Transformers to extract key points.



###⚠️ **Warning:** After you run the cell, **wait a few seconds** ⏳ for the UI to show up.

Try it now — paste, summarize, done. 🎯


```python
!pip install -q gradio transformers

from transformers import pipeline
import gradio as gr

# this is the Model used to Summarizer
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# handel the input and get the Summarizer text
def summarize(text):
    if not text.strip():
        return "⚠️ There is no text."
    result = summarizer(text, max_length=130, min_length=30, do_sample=False)
    return result[0]['summary_text']

# Gradio GUI
gr.Interface(
    fn=summarize,
    inputs=gr.Textbox(lines=10, placeholder="Paste your English text here..."),
    outputs=gr.Textbox(label="Summary"),
    title="🧠 Text Summarizer",
    description="Paste your text and get summary using a BART model.",
    theme="default",
).launch(share=True)
```
